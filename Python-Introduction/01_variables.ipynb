## Tuples

A tuple is an ordered, immutable collection of items.

Tuples are often used for:
- fixed collections of values (e.g. coordinates)
- returning multiple values from a function

```python
coordinates = (50.087, 14.421)

print(coordinates[0])

# This will raise an error
coordinates[0] = 51.0
```

## Mutable and immutable data types
- list = mutable (we can do changes on already created list)
- tuple = immutable (we cannot do changes on already created tuple)

```python
# this works for tuple because we are creating a new one
t = (1, 2, 3)
t = t + (4,)

print(t)
```

```python
# assigning list a to variable b
a = [1, 2, 3]
b = a

print(b)

a.append(4)

print(b)
```

```python
# assigning tuple a to variable b
a = (1, 2, 3)
b = a

print(b)

a = a + (4,)

print(b)
```

## Dictionaries

A dictionary stores key–value pairs
- access values using keys
- add or update values

Dictionaries are commonly used for:
- structured data
- records (rows in a table)
- JSON-like data

Based on the key we can easily look for the value

```python
person = {
'name': 'Alice',
'age': 30,
'city': 'Prague'
}

print(person['name'])
print(person['age'])

#TODO: change the age to 31


#adding new key and value:
person['email'] = 'alice@example.com'
print(person)

#TODO: add new key 'siblings' what type of values should we expect? choose data type and structure accordingly and assign the key some value
```
