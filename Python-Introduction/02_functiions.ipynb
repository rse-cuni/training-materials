# Functions

Functions are reusable blocks of code that perform a specific task. They help us avoid repeating code and make our programs more organized.

## Defining a function

We use the `def` keyword to define a function.

## Function parameters

Parameters are variables that we can pass into a function.

## Return values

Functions can return values using the `return` statement.

````python
# Simple function without parameters
def greet():
    print("Hello, welcome!")

greet()  # Calling the function
````

````python
# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Alice")
greet_person("Bob")
````

````python
# Function with return value
def add_numbers(a, b):
    result = a + b
    return result

sum_result = add_numbers(5, 3)
print(f"Sum: {sum_result}")
````

## Built-in Functions

Python comes with many built-in functions that we can use right away.

### len() - Get length

Returns the number of items in an object.

````python
# Using len() with different data types
text = "Hello"
my_list = [1, 2, 3, 4, 5]

print(f"Length of text '{text}': {len(text)}")
print(f"Length of list {my_list}: {len(my_list)}")
````

### sum() - Sum values

Returns the sum of all items in an iterable.

````python
# Using sum() with lists
numbers = [1, 2, 3, 4, 5]
print(f"Sum of {numbers}: {sum(numbers)}")
````

### min() and max() - Find minimum and maximum

Find the smallest and largest values in an iterable.

````python
# Using min() and max()
numbers = [10, 5, 8, 20, 3]
print(f"Numbers: {numbers}")

minimum = min(numbers)
print(f"Minimum: {minimum}")

#TODO: print the maximum value in the list
````

### sorted() - Sort items

Returns a new sorted list from the items in an iterable.

````python
# Using sorted()
unsorted_list = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"Unsorted: {unsorted_list}")
print(f"Sorted: {sorted(unsorted_list)}")

# Sorting in descending order
print(f"Sorted descending: {sorted(unsorted_list, reverse=True)}")
````
