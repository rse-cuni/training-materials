## Conditions - if, elif, else

Conditions allow us to make decisions in our code based on certain criteria.

### Comparison Operators

- `==` : equal to
- `!=` : not equal to
- `>` : greater than
- `<` : less than
- `>=` : greater than or equal to
- `<=` : less than or equal to

### Logical Operators

- `and` : both conditions must be true
- `or` : at least one condition must be true
- `not` : negates a condition

Let's see some examples:

```python
# Basic if statement
age = 18
if age >= 18:
    print("You are an adult")
```

```python
#if-else statement
temperature = 25
if temperature > 30:
    print("It's hot today")
else:
    print("It's not too hot")

#TODO: Change temperature to 32 and see what happens
```

```python
# if-elif-else statement
score = 85
if score >= 90 and score <= 100:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"Your grade is {grade}")
```

## Loops - for and while

Loops allow us to repeat code multiple times.

### for loop

Use when you know how many times to iterate or want to iterate over a sequence.

```python
# Loop through a list
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(f"I like {fruit}")

# Loop using range()
for i in range(5):
    print(f"Count: {i}")

# TODO: Print numbers 1 to 10 using for loop
```

## Working with Files

Python allows us to read from and write to files.

### Opening and closing files

We can open files using the `open()` function and then close them with `close()`.

```python
# Writing to a file (manual close)
file = open('example.txt', 'w')
file.write('Hello, this is a test file!\n')
file.write('This is the second line.\n')
file.close()

# Reading from a file (manual close)
file = open('example.txt', 'r')
content = file.read()
print(content)
file.close()
```

### Using 'with' statement (recommended)

The `with` statement automatically closes files for us, even if an error occurs.

```python
# Writing with 'with' statement
with open('example2.txt', 'w') as file:
    file.write('Hello, this is written with the with statement!\n')
    file.write('This is the second line.\n')

# Reading with 'with' statement
with open('example2.txt', 'r') as file:
    content = file.read()
    print(content)
```

## NumPy

NumPy (Numerical Python) is a powerful library for working with arrays and mathematical operations.

### Arrays vs Lists

NumPy arrays are more efficient and support vectorized operations compared to Python lists.

```python
# Creating arrays
import numpy as np

# Creating a list
my_list = [1, 2, 3, 4, 5]

# Creating a NumPy array
my_array = np.array([1, 2, 3, 4, 5])

print(f"List: {my_list}")
print(f"Array: {my_array}")

# Operations on arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([6, 7, 8, 9, 10])

# Element-wise operations
print(f"Addition: {arr1 + arr2}")
print(f"Multiplication: {arr1 * 2}")
print(f"Square root: {np.sqrt(arr1)}")
```
