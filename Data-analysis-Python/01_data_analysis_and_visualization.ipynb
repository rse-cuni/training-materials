```python
# Exercise 1: Histogram of salaries
plt.figure(figsize=(8, 6))
plt.hist(df['salary'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Exercise 2: Scatter plot of age vs salary
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['salary'], color='red')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Exercise 3: Bar chart of average salary by department
dept_avg_salary = df.groupby('department')['salary'].mean()
plt.figure(figsize=(8, 6))
plt.bar(dept_avg_salary.index, dept_avg_salary.values, color='green')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Exercise 4: Box plot of salaries by department
plt.figure(figsize=(8, 6))
df.boxplot(column='salary', by='department', ax=plt.gca())
plt.title('Salary Distribution by Department')
plt.suptitle('')  # Remove automatic title
plt.show()
```

### Mini-project Exercise

Let's work with a more complex dataset. We'll create a synthetic dataset of sales data to practice our skills.

1. Create a synthetic sales dataset
2. Load and explore the data
3. Clean the data
4. Perform calculations
5. Create visualizations

```python
# Create synthetic sales dataset
np.random.seed(123)
dates = pd.date_range('2023-01-01', periods=1000, freq='D')
products = ['Product A', 'Product B', 'Product C', 'Product D']
regions = ['North', 'South', 'East', 'West']

sales_data = {
    'date': np.random.choice(dates, 1000),
    'product': np.random.choice(products, 1000),
    'region': np.random.choice(regions, 1000),
    'sales_amount': np.random.uniform(100, 1000, 1000),
    'units_sold': np.random.randint(1, 50, 1000)
}

sales_df = pd.DataFrame(sales_data)
sales_df.head(10)
```

```python
# Exercise: Load and explore the data
print('Dataset shape:', sales_df.shape)
print('\nDataset info:')
sales_df.info()
print('\nDataset description:')
sales_df.describe()
```

```python
# Exercise: Clean the data
# Check for missing values
print('Missing values:')
sales_df.isnull().sum()

# Check for duplicates
print('\nNumber of duplicates:', sales_df.duplicated().sum())

# Remove duplicates if any
sales_df = sales_df.drop_duplicates()
print('\nDataset shape after removing duplicates:', sales_df.shape)
```

```python
# Exercise: Perform calculations
# Total sales by product
total_sales_by_product = sales_df.groupby('product')['sales_amount'].sum()
print('Total sales by product:')
print(total_sales_by_product)

# Average sales by region
avg_sales_by_region = sales_df.groupby('region')['sales_amount'].mean()
print('\nAverage sales by region:')
print(avg_sales_by_region)

# Sales per unit for each product
sales_per_unit = sales_df.groupby('product')['sales_amount'].sum() / sales_df.groupby('product')['units_sold'].sum()
print('\nSales per unit by product:')
print(sales_per_unit)
```

```python
# Exercise: Create visualizations
# 1. Sales trend over time
sales_by_date = sales_df.groupby('date')['sales_amount'].sum()
plt.figure(figsize=(12, 6))
plt.plot(sales_by_date.index, sales_by_date.values)
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Sales by product
plt.figure(figsize=(8, 6))
plt.bar(total_sales_by_product.index, total_sales_by_product.values, color='orange')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
